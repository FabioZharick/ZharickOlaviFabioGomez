```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = 'MisDatosDerivadaCentral.txt'
data = np.loadtxt(file)
X = data[:, 0]
Y = data[:, 1]

plt.plot(X,Y, label = r'd$e^{x^{2}}$/dx')
plt.title('Derivada central empleando c++')
plt.xlabel('x')
plt.ylabel('df/dx')
plt.legend()
plt.show()
```
